```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
data=pd.read_csv("Sample - Superstore.csv",encoding="windows-1252")
data.head()
```

```python
# Sales distribution by category
category_sales = data.groupby('Category').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()

# Sales distribution by region
region_sales = data.groupby('Region').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()

# Display the results
print(category_sales)
print(region_sales)
```

```python
import matplotlib.pyplot as plt

# Convert 'Order Date' to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Set the 'Order Date' as the index
data.set_index('Order Date', inplace=True)

# Resample by month and calculate total sales
monthly_sales = data['Sales'].resample('M').sum()

# Plot monthly sales trends
plt.figure(figsize=(10, 6))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.show()
```

```python
# Total sales by customer
customer_sales = data.groupby('Customer ID').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Orders=('Order ID', 'nunique')
).reset_index()

# Display the top 10 customers based on total sales
top_customers = customer_sales.sort_values(by='Total_Sales', ascending=False).head(10)
print(top_customers)
```

```python
# Top products by sales and quantity
product_sales = data.groupby('Product Name').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Quantity=('Quantity', 'sum')
).reset_index()

# Top 10 products by sales
top_products_sales = product_sales.sort_values(by='Total_Sales', ascending=False).head(10)

# Top 10 products by quantity
top_products_quantity = product_sales.sort_values(by='Total_Quantity', ascending=False).head(10)

print(top_products_sales)
print(top_products_quantity)
```

```python
# Bar plot for Sales by Category
plt.figure(figsize=(8,6))
sns.barplot(x='Category', y='Total_Sales', data=category_sales, palette='Blues_d')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Bar plot for Profit by Category
plt.figure(figsize=(8,6))
sns.barplot(x='Category', y='Total_Profit', data=category_sales, palette='Greens_d')
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.show()
```

```python
# Bar plot for Sales by Region
plt.figure(figsize=(8,6))
sns.barplot(x='Region', y='Total_Sales', data=region_sales, palette='Purples_d')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Bar plot for Profit by Region
plt.figure(figsize=(8,6))
sns.barplot(x='Region', y='Total_Profit', data=region_sales, palette='Blues_d')
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.figure(figsize=(7, 7))
region_sales_pie = region_sales.set_index('Region')['Total_Sales']
region_sales_pie.plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set2', legend=False)
plt.title('Sales Distribution by Region')
plt.ylabel('')
plt.show()
```

```python
category_sales.set_index('Category')[['Total_Sales', 'Total_Profit']].plot(kind='bar', stacked=True)
plt.title('Total Sales and Profit by Category')
plt.ylabel('Amount')
plt.show()
```
